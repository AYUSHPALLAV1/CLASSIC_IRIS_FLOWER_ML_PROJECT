```python
def plot_incorrect_prediction(df_prediction,x_axis_feature,y_axis_features):
    fig,axs=plt.subplots(2,2,figsize=(10,10))
    axs=axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_features,hue="prediction_label",data=df_prediction,ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_features,hue="target_names",data=df_prediction,ax=axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_features,hue="correct_prediction",data=df_prediction,ax=axs[2])
    axs[3].set_visible(False)
    plt.show()


    
```

```python
plot_incorrect_prediction(df_predictions,"petal length (cm)","petal width (cm)")
```
